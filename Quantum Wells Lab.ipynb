```python
### Question 2 ###
import numpy as np
import scipy.optimize as sp
import matplotlib.pyplot as plt

EstimatedIntersection1 = np.array([1,3.5])
EstimatedIntersection2 = 2.2
array = np.linspace(0,8,10000)

def q(x):
    return np.sqrt(16-x**2)
def p(x):
    return x*np.tan(x)
def r(x):
    return -1 * x * (np.tan(x))**-1

def FuncToSolve1(x):
     return q(x) - p(x)
def FuncToSolve2(x):
    return q(x) - r(x)

solution1 = sp.fsolve(FuncToSolve1,EstimatedIntersection1)
solution2 = sp.fsolve(FuncToSolve2,EstimatedIntersection2)
print(solution1,solution2)


plt.plot(array,q(array),label='q(x)')
plt.plot(array,p(array),label='p(x)')
plt.plot(array,r(array),label='r(x)')
plt.legend()
plt.xlim(0,8)
plt.ylim(0,5)
plt.show()
```

```python
hbar = 1.05457e-34
Bar_X = 0.32
QW_X = 0
Eg_QW = 1.424
Eg_Bar = 1.823
a = np.linspace(1e-9,1e-8,10)
x = np.linspace(0,20,1000)
guesse = np.array([1,1.1,1.2,1.2,1.3,1.4,1.4,1.4,1.4,1.4])
guessh = np.array([1.2,1.3,1.4,1.4,1.4,1.4,1.4,1.4,1.4,1.4])

h = 6.63e-34
m_0 = 9.11e-31
m_e = 0.067*m_0
m_h = 0.45*m_0
#https://ecee.colorado.edu/~bart/book/effmass.htm

Ve_0 = (0.67*(Eg_Bar-Eg_QW))
Vh_0 = (0.33*(Eg_Bar-Eg_QW))

electronE = np.zeros(len(a))
holeE = np.zeros(len(a))
TotalE = np.zeros(len(a))

for i in range(len(a)):
    
    E0e = (h**2)/(8*m_e*a[i]**2)/1.6e-19
    
    def func_e(x):
        return (np.sqrt((Rel_de*np.pi**2)/4-x**2) - x*np.tan(x))
    
    Rel_de = ((Ve_0*8*m_e*(a[i])**2)*1.6e-19/(hbar**2*4*np.pi**2))

    sol_e = sp.fsolve(func_e,guesse[i])
    
    electronE[i] = ((4*sol_e**2)/(np.pi**2))*E0e

    E0h = (h**2)/(8*m_h*a[i]**2)/1.6e-19

    
    def func_h(x):
        return (np.sqrt((Rel_dh*np.pi**2)/4-x**2) - x*np.tan(x))
    
    Rel_dh = (Vh_0*8*m_h*(a[i])**2)*1.6e-19/(hbar**2*4*np.pi**2)
    
    sol_h = sp.fsolve(func_h,guessh[i])
    
    holeE[i] = ((4*sol_h**2)/(np.pi**2))*E0h
    
    TotalE[i] = Eg_QW + electronE[i] + holeE[i]
    

emissE = ((h*3e8)/815e-9)/1.6e-19
plt.hlines(emissE,1e-9,1e-8,color = 'r')
plt.plot(a,TotalE)
plt.title('Energy as a function of quantum well width')
plt.xlim(0.51e-8,0.52e-8)
plt.ylim(1.52,1.53)
plt.xlabel('Quantum well width (m)')
plt.ylabel('E0(a) (ev)')
plt.show()
##qw well width = 5.15e-9m
```

```python
###wavefunction 5C

WaveArray = np.linspace(-10e-9,10e-9,1000)
energy1 = 1.30496504
energy2 = 2.59249957
A = 8.52e-9
V0 = 0.52*1.6e-19

def g17(WaveArray,energy1,A,e_mass,hbar,V0):

    g17graph = np.zeros(len(WaveArray))
    k1 = (2/A)*energy1
    k2 = (2/A)*(np.sqrt((e_mass*(V0)*A**2)/(2*hbar**2)-energy1**2))
    
    for i in range(len(WaveArray)):
        
        if WaveArray[i] < (-A/2):
            g17graph[i] = np.cos(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(k2*WaveArray[i]))
            
        elif (-A/2) < WaveArray[i] < (A/2):
            g17graph[i] = np.cos(k1*WaveArray[i])
            
        else:
            g17graph[i] = np.cos(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp((-k2)*WaveArray[i]))
        
    return g17graph

def g19(WaveArray,energy2,A,e_mass,hbar,V0):
    
    g19graph = np.zeros(len(WaveArray))
    k1 = (2/A)*energy2
    k2 = (2/A)*(np.sqrt((e_mass*(V0)*A**2)/(2*hbar**2)-energy2**2))
    
    for i in range(len(WaveArray)):
        
        if WaveArray[i] <= (-A/2):
            g19graph[i] = -np.sin(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(k2*WaveArray[i]))
            
        elif (-A/2) <= WaveArray[i] <= (A/2):
            g19graph[i] = np.sin(k1*WaveArray[i])
            
        else:
            g19graph[i] = np.sin(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(-k2*WaveArray[i]))
    
    return g19graph

fig, ((ax1),(ax2)) = plt.subplots(2,1,figsize=(5,7))

g17graph = g17(WaveArray,energy1,A,e_mass,hbar,V0)
g19graph = g19(WaveArray,energy2,A,e_mass,hbar,V0)
ax1.plot(WaveArray,g17graph,label = 'Ground State')
ax1.plot(WaveArray,g19graph, label = 'First Excited State')
ax1.set_title('Eigenfunctions for the Bound Eigenstates')
ax1.vlines(-A/2,-1.1,1.1,color='crimson',label = 'QM "Well Walls"')
ax1.vlines(A/2,-1.1,1.1,color='crimson')
ax1.set_ylim(-1.1,1.1)
ax1.legend()
ax2.plot(WaveArray,(abs(g17graph))**2)
ax2.plot(WaveArray,(abs(g19graph))**2)
ax2.vlines(-A/2,0,1.1,color='crimson')
ax2.vlines(A/2,0,1.1,color='crimson')
ax2.set_ylim(0,1.1)
ax2.set_title('Probability Density Wave Function')
plt.xlabel('X')
plt.tight_layout()
plt.plot()
%time
```
