```python
###wavefunction 5C

WaveArray = np.linspace(-2e-8,2e-8,1000)
energy1 = 1.13908085
energy2 = 2.20036538
A = 7.16e-9
V0 = 0.52*1.6e-19

def g17(WaveArray,energy1,A,e_mass,hbar,V0):

    g17graph = np.zeros(len(WaveArray))
    k1 = (2/A)*energy1
    k2 = (2/A)*(np.sqrt((e_mass*(V0)*A**2)/(2*hbar**2)-energy1**2))
    
    for i in range(len(WaveArray)):
        
        if WaveArray[i] < (-A/2):
            g17graph[i] = np.cos(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(k2*WaveArray[i]))
            
        elif (-A/2) < WaveArray[i] < (A/2):
            g17graph[i] = np.cos(k1*WaveArray[i])
            
        else:
            g17graph[i] = np.cos(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp((-k2)*WaveArray[i]))
        
    return g17graph

def g19(WaveArray,energy2,A,e_mass,hbar,V0):
    
    g19graph = np.zeros(len(WaveArray))
    k1 = (2/A)*energy2
    k2 = (2/A)*(np.sqrt((e_mass*(V0)*A**2)/(2*hbar**2)-energy2**2))
    
    for i in range(len(WaveArray)):
        
        if WaveArray[i] <= (-A/2):
            g19graph[i] = -np.sin(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(k2*WaveArray[i]))
            
        elif (-A/2) <= WaveArray[i] <= (A/2):
            g19graph[i] = np.sin(k1*WaveArray[i])
            
        else:
            g19graph[i] = np.sin(k1*(A/2))*(np.exp(k2*(A/2)))*(np.exp(-k2*WaveArray[i]))
    
    return g19graph

fig, ((ax1),(ax2)) = plt.subplots(1,2,figsize=(20,7))

g17graph = g17(WaveArray,energy1,A,e_mass,hbar,V0)
g19graph = g19(WaveArray,energy2,A,e_mass,hbar,V0)
ax1.plot((WaveArray/1e-8),(g17graph),label = r'$\Psi_1(x)$',color='k')
ax1.plot(WaveArray/1e-8,g19graph, label = r'$\Psi_2(x)$',color='b')
ax1.set_ylim(0,1.1)
ax1.set_xlim(-1.5,1.5)
ax1.text(-1.45,1.02,r'(a)',fontsize='x-large')

ax1.vlines((-A/2)/1e-8,-1.1,1.1,color='crimson',label = '"Well Walls"',linestyle='dashdot')
ax1.vlines((A/2)/1e-8,-1.1,1.1,color='crimson',linestyle='dashdot')
ax1.set_ylim(-1.1,1.1)
ax1.set_xlim(-1.5,1.5)
ax1.legend()
ax1.set_xlabel('X',fontsize = 'x-large')
ax2.plot((WaveArray/1e-8),(abs(g17graph))**2,color='k')
ax2.plot((WaveArray/1e-8),(abs(g19graph))**2,color='b')
ax2.vlines((-A/2)/1e-8,0,1.1,color='crimson',linestyle='dashdot')
ax2.vlines((A/2)/1e-8,0,1.1,color='crimson',linestyle='dashdot')
ax2.set_ylim(0,1.1)
ax2.set_xlim(-1.5,1.5)
ax2.set_xlabel('X',fontsize = 'x-large')
ax2.text(-1.45,1.05,r'(b)',fontsize='x-large')

plt.tight_layout()
plt.savefig(r'C:\Users\anton\OneDrive\Desktop\UNI\Third Year\PH 390 Experimental Physics\QW Lab\savefig\question5c.jpeg',dpi=500)
plt.plot()
%time
```
